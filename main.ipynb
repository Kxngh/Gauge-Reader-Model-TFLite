```python
import os

image_dir = "gauge_image_dataset"

file_sizes = {}
duplicates = []

for filename in os.listdir(image_dir):
    filepath = os.path.join(image_dir, filename)
    try:
        file_size = os.path.getsize(filepath)
        if file_size in file_sizes:
            duplicates.append(filepath)
        else:
            file_sizes[file_size] = filepath
    except Exception as e:
        print(f"Error processing {filepath}: {e}")

for duplicate in duplicates:
    os.remove(duplicate)
    print(f"Removed duplicate: {duplicate}")

print("Duplicate removal complete.")
```

```python
import os
import tensorflow as tf
from PIL import Image
import numpy as np

# Paths
input_folder = "gauge_image_dataset"
output_folder = "augmented_gauge_dataset"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Augmentation function
def augment_image(image):
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_flip_up_down(image)
    image = tf.image.random_brightness(image, max_delta=0.2)
    image = tf.image.random_contrast(image, lower=0.5, upper=1.5)
    image = tf.image.random_saturation(image, lower=0.5, upper=1.5)
    return image

# Process images in the input folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, f"aug_{filename}")
        
        # Open the image and convert to RGB (discard alpha channel)
        image = Image.open(input_path).convert("RGB")
        image_array = tf.convert_to_tensor(np.array(image), dtype=tf.float32)
        image_array = tf.expand_dims(image_array, 0)  # Add batch dimension

        # Apply augmentation
        augmented_image = augment_image(image_array[0])  # Remove batch dimension
        augmented_image = tf.cast(augmented_image, tf.uint8)  # Convert back to uint8

        # Convert back to PIL and save
        augmented_image_pil = Image.fromarray(augmented_image.numpy())
        augmented_image_pil.save(output_path)

        print(f"Augmented image saved: {output_path}")

print("Data augmentation complete!")
```

```python
import cv2
import numpy as np

def generate_gauge(radius=100, angle=45, center=(150, 150)):
    image = np.zeros((300, 300, 3), dtype=np.uint8)

    cv2.circle(image, center, radius, (255, 255, 255), 2)

    cv2.circle(image, center, 3, (0, 255, 0), -1)  
    cv2.line(image, (center[0] - radius, center[1]), (center[0] + radius, center[1]), (255, 255, 255), 1)

    needle_length = int(radius * 0.9)
    needle_x = int(center[0] + needle_length * np.cos(np.radians(angle)))
    needle_y = int(center[1] - needle_length * np.sin(np.radians(angle)))
    cv2.line(image, center, (needle_x, needle_y), (0, 0, 255), 2)

    return image

for i in range(100):
    angle = np.random.randint(0, 360)  
    gauge = generate_gauge(angle=angle)
    cv2.imwrite(f"synthetic_gauge_{i}.png", gauge)
```

```python
import os
import math
import xml.etree.ElementTree as ET

annotation_dir = "valid"

for filename in os.listdir(annotation_dir):
    if filename.lower().endswith(".xml"):
        filepath = os.path.join(annotation_dir, filename)
        
        # Parse the XML file
        tree = ET.parse(filepath)
        root = tree.getroot()

        needle_base_coords = None
        needle_tip_coords = None

        # Iterate through all 'object' elements
        for obj in root.findall('object'):
            name_el = obj.find('name')
            if name_el is None:
                continue
            obj_name = name_el.text.strip()
            
            bndbox = obj.find('bndbox')
            if bndbox is None:
                continue
            
            # Extract bounding box coordinates
            xmin = float(bndbox.find('xmin').text)
            xmax = float(bndbox.find('xmax').text)
            ymin = float(bndbox.find('ymin').text)
            ymax = float(bndbox.find('ymax').text)
            
            # Compute center of bounding box
            cx = (xmin + xmax) / 2.0
            cy = (ymin + ymax) / 2.0
            
            # Assign coordinates based on object name
            if obj_name == 'Needle_base':
                needle_base_coords = (cx, cy)
            elif obj_name == 'Needle_Tip':
                needle_tip_coords = (cx, cy)
        
        # Check if both coordinates were found
        if needle_base_coords is None or needle_tip_coords is None:
            print(f"Skipping {filename}: Missing 'Needle_base' or 'Needle_Tip' objects.")
            continue

        # Compute angle
        (bx, by) = needle_base_coords
        (tx, ty) = needle_tip_coords

        dx = tx - bx
        dy = ty - by

        # Using the same angle convention: 
        # 0 degrees at vertical (needle pointing up), increasing clockwise.
        raw_angle_radians = math.atan2(dx, -dy)
        angle_degrees = math.degrees(raw_angle_radians)
        if angle_degrees < 0:
            angle_degrees += 360.0

        # Add/update needle_angle_degrees element in XML
        angle_el = root.find('needle_angle_degrees')
        if angle_el is None:
            angle_el = ET.Element('needle_angle_degrees')
            root.append(angle_el)
        angle_el.text = f"{angle_degrees:.2f}"

        # Write back updated XML
        tree.write(filepath, encoding='utf-8', xml_declaration=True)

        print(f"Updated {filename} with angle {angle_degrees:.2f}")
```

```python
# DataLoader Class
class DataLoader:
    def __init__(self, image_dir, label_dir, img_size):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.img_size = img_size

    def load_data(self):
        images = []
        labels = []

        image_files = glob(os.path.join(self.image_dir, "*.jpg"))
        for img_file in image_files:
            # Load image
            img = cv2.imread(img_file)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, (self.img_size, self.img_size)) / 255.0
            images.append(img)

            # Load corresponding label
            label_file = os.path.join(self.label_dir, os.path.basename(img_file).replace('.jpg', '.txt'))
            with open(label_file, 'r') as f:
                label = np.array([list(map(float, line.strip().split())) for line in f.readlines()])
            labels.append(label)

        return np.array(images), np.array(labels)

# Instantiate DataLoader
img_size = 416
train_loader = DataLoader("./train/images", "./train/labels", img_size)
val_loader = DataLoader("./val/images", "./val/labels", img_size)

x_train, y_train = train_loader.load_data()
x_val, y_val = val_loader.load_data()

print("Data loaded successfully.")
```

```python
# Define EfficientDet Model
base_model = EfficientNetB0(include_top=False, input_tensor=Input(shape=(img_size, img_size, 3)))
x = Flatten()(base_model.output)
x = Dense(256, activation="relu")(x)
x = Dropout(0.5)(x)
output = Dense(len(y_train[0]), activation="sigmoid")(x)

model = Model(inputs=base_model.input, outputs=output)

print("Model defined successfully.")

# Compile Model
model.compile(optimizer=Adam(learning_rate=1e-4), 
              loss="sparse_categorical_crossentropy", 
              metrics=["accuracy"])
print("Model compiled successfully.")

# Callbacks
early_stopping = EarlyStopping(monitor="val_loss", patience=10, restore_best_weights=True)
model_checkpoint = ModelCheckpoint("efficientdet_model.h5", save_best_only=True, monitor="val_loss", mode="min")

# Train Model
history = model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    epochs=50,
    batch_size=32,
    callbacks=[early_stopping, model_checkpoint],
    verbose=1
)

print("Model training completed.")
```

```python
# Save Trained Model
model.save("efficientdet_trained_model.h5")
print("Model saved successfully.")
```
